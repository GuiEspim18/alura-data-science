```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canadian_immegration_data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.set_index("Country", inplace=True)
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
years
```

```python
brazil = df.loc['Brazil', years]
```

```python
brazil
```

```python
brazil_dict = {"year": brazil.index.tolist(), "immigrants": brazil.values.tolist()}
data_brazil = pd.DataFrame(brazil_dict)
```

```python
data_brazil
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data_brazil['year'], data_brazil['immigrants'])
plt.title("Immigration from Brazil to Canada")
plt.xlabel("year")
plt.ylabel("Immigrants quantity")
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show() # tudo o que eu digitiar após o show nada vai ser executado
```
